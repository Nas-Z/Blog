```python
import requests
import json
from newspaper import Article #library for scrapping news article in a specific URL
import urllib3 #for dissabling exceptions warnings messages
```

```python
response = requests.get(custom_search_API)
```

```python
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning) # to not print InsecureRequestWarning for each url request
if(response.status_code == 200):
  data = response.json()
  returned_data = {}
  for item in data["news_results"][:9]:
    returned_data[item['position']] = {'title': item['title'], 'url': item['link'], 'source': item["source"]["name"]}

    url = item['link']
    article_response = requests.get(url, verify=False)
    if article_response.status_code == 200:
      article = Article(url)
      article.set_html(article_response.text)
      article.parse()
      returned_data[item['position']]['content'] = article.text

  print(json.dumps(returned_data, ensure_ascii=False ,indent = 2))
  # print(returned_data["1"]["content"])
else:
   print(f"Error: {response.status_code}, {response.text}")
```
